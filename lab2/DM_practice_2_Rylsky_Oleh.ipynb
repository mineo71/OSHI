```python
inner_radius = 5
inner_width = 1
inner_points = 500

outer_radius = 10
outer_width = 1
outer_points = 500
```

```python
def generate_torus_points(radius, width, num_points):
    angles = np.random.uniform(0, 2 * np.pi, num_points)
    radii = np.random.uniform(radius - width / 2, radius + width / 2, num_points)
    x = radii * np.cos(angles)
    y = radii * np.sin(angles)
    return x, y
```

```python
inner_x, inner_y = generate_torus_points(inner_radius, inner_width, inner_points)
inner_labels = np.ones(inner_points)

outer_x, outer_y = generate_torus_points(outer_radius, outer_width, outer_points)
outer_labels = np.full(outer_points, 2)
```

```python
x = np.concatenate([inner_x, outer_x])
y = np.concatenate([inner_y, outer_y])
labels = np.concatenate([inner_labels, outer_labels])

data = pd.DataFrame({'x': x, 'y': y, 'class': labels})
data.to_csv('torus_data.csv', index=False)
```

```python
plt.figure(figsize=(8, 8))
plt.scatter(inner_x, inner_y, c='green', label='Class 1')
plt.scatter(outer_x, outer_y, c='pink', label='Class 2')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Torus in Torus')
plt.show()

inner_radii = np.sqrt(inner_x**2 + inner_y**2)
outer_radii = np.sqrt(outer_x**2 + outer_y**2)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(inner_radii, bins=30, color='green', alpha=0.7)
plt.title('Inner Torus Radii Distribution')

plt.subplot(1, 2, 2)
plt.hist(outer_radii, bins=30, color='pink', alpha=0.7)
plt.title('Outer Torus Radii Distribution')

plt.show()
```

* For each torus, generate random angles and radii within the given parameters.
* Calculate the points using trigonometric functions.

* Data is saved to a CSV file.

### 5.1: Теоретичне підґрунтя

**Опис системи Лоренца:**
Система Лоренца — це набір трьох нелінійних диференціальних рівнянь, що описують динаміку тривимірної системи з хаотичною поведінкою. Вона визначається такими рівняннями:

$$
\frac{dx}{dt} = \sigma (y - x)
$$

$$
\frac{dy}{dt} = x (\rho - z) - y
$$

$$
\frac{dz}{dt} = xy - \beta z
$$

Де:
- $(x, y, z)$ — змінні стану системи.
- $\sigma$ — параметр, що відповідає за швидкість зміни (часто називають "параметр Прандтля").
- $\rho$ — параметр, що відповідає за різницю температур у системі (відповідає за інтенсивність конвекції).
- $\beta$  — параметр, що відповідає за співвідношення розмірів.

**Роз'яснення до завдання:**
- Згенерувати дані для системи Лоренца за допомогою різних значень параметрів $\sigma$, $\rho$ та $\beta$.
- Проаналізувати вплив зміни параметрів на траєкторії системи.
- Візуалізувати результати в тривимірному просторі для кожного набору даних.

### 5.2: Генерація та візуалізація набору даних

```python
def lorenz(t, state, sigma, rho, beta):
    x, y, z = state
    dxdt = sigma * (y - x)
    dydt = x * (rho - z) - y
    dzdt = x * y - beta * z
    return [dxdt, dydt, dzdt]
```

```python
params = [
    (10, 28, 8/3),
    (14, 46.92, 4),
    (20, 35, 2)
]
```

```python
initial_state = [1.0, 1.0, 1.0]
t_span = (0, 50)
t_eval = np.linspace(t_span[0], t_span[1], 10000)
```

```python
from scipy.integrate import solve_ivp

fig = plt.figure(figsize=(18, 6))

for i, (sigma, rho, beta) in enumerate(params):
    sol = solve_ivp(lorenz, t_span, initial_state, args=(sigma, rho, beta), t_eval=t_eval)
    x, y, z = sol.y

    ax = fig.add_subplot(1, 3, i+1, projection='3d')
    ax.plot(x, y, z, lw=0.5)
    ax.set_title(f'σ={sigma}, ρ={rho}, β={beta}')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')

plt.tight_layout()
plt.show()
```

Step:
* Select several combinations of parameters
